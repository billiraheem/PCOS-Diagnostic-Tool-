```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.width', None)        # No line wrapping
plt.style.use('seaborn-v0_8-whitegrid')     # Clean plot style

print("Libraries imported successfully!")
print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")
```

```python
# Load the PCOS Dataset
data_path = "../../Dataset3/data without infertility _final.csv"

# Load the data
df = pd.read_csv(data_path)

# Basic info
print(f"Dataset Shape: {df.shape[0]} rows × {df.shape[1]} columns")
print(f"\nTarget Column: 'PCOS (Y/N)'")
print(f"\nFirst 5 rows:")
df.head()
```

```python
# View All Column Names
print("All Columns in Dataset:\n")
for i, col in enumerate(df.columns, 1):
    print(f"{i:2}. {col}")
```

```python
# Data Types & Basic Info
print("Data Types & Non-Null Counts:\n")
df.info()
```

```python
# Statistical Summary
print("Statistical Summary (Numerical Columns):\n")
df.describe()
```

```python
# Missing Values Analysis
print("Missing Values per Column:\n")

missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_pct
})

# Show only columns with missing values
missing_with_nulls = missing_df[missing_df['Missing Count'] > 0]
if len(missing_with_nulls) > 0:
    print(missing_with_nulls.sort_values('Missing Count', ascending=False))
else:
    print("No missing values found!")
    
print(f"\nTotal Missing Values: {missing.sum()}")
```

```python
# Investigate AMH Column (why is it 'object' type?)
# Show unique values that can't be converted to numbers
amh_col = df['AMH(ng/mL)']

# Try to convert to numeric - errors='coerce' turns failures into NaN
amh_numeric = pd.to_numeric(amh_col, errors='coerce')

# Find rows where conversion failed (became NaN but original wasn't null)
problematic = df[amh_numeric.isna() & amh_col.notna()]['AMH(ng/mL)']

print("Problematic AMH values that can't be converted to numbers:\n")
print(problematic.unique())
print(f"\nCount of problematic rows: {len(problematic)}")
```

```python
# Check what's in the Unnamed column
print("Unnamed:42 column values:\n")
print(df['Unnamed: 42'].value_counts(dropna=False))
```

```python
# Data Cleaning
# Make a copy to preserve original
df_clean = df.copy()

# Step 1: Drop useless columns
columns_to_drop = ['Unnamed: 42', 'Sl. No', 'Patient File No.']
df_clean = df_clean.drop(columns=columns_to_drop)
print(f"Dropped columns: {columns_to_drop}")

# Step 2: Fix AMH column - convert to numeric (errors become NaN)
df_clean['AMH(ng/mL)'] = pd.to_numeric(df_clean['AMH(ng/mL)'], errors='coerce')
print(f"Converted AMH to numeric (invalid values → NaN)")

# Step 3: Check remaining missing values
print(f"\nMissing values after cleaning:")
missing = df_clean.isnull().sum()
missing_cols = missing[missing > 0]
print(missing_cols)
```

```python
# Impute Missing Values
# Find columns with missing values
missing_cols = df_clean.columns[df_clean.isnull().any()].tolist()
print("Columns with missing values:", missing_cols)

for col in missing_cols:
    missing_count = df_clean[col].isnull().sum()
    
    if df_clean[col].dtype in ['int64', 'float64']:
        # Numerical: use median
        median_val = df_clean[col].median()
        df_clean[col] = df_clean[col].fillna(median_val)
        print(f"{col}: filled {missing_count} values with median = {median_val:.2f}")
    else:
        # Categorical: use mode (most frequent)
        mode_val = df_clean[col].mode()[0]
        df_clean[col] = df_clean[col].fillna(mode_val)
        print(f"{col}: filled {missing_count} values with mode = {mode_val}")

# Verify no more missing values
print(f"\nTotal missing values now: {df_clean.isnull().sum().sum()}")
print(f"Clean dataset shape: {df_clean.shape}")
```

```python
# Target Variable Analysis (Class Distribution)
target_col = 'PCOS (Y/N)'

# Count of each class
class_counts = df_clean[target_col].value_counts()
class_pct = (df_clean[target_col].value_counts(normalize=True) * 100).round(2)

print("Target Variable Distribution:\n")
print(f"0 (No PCOS):  {class_counts[0]} patients ({class_pct[0]}%)")
print(f"1 (Has PCOS): {class_counts[1]} patients ({class_pct[1]}%)")

# Visualize
fig, ax = plt.subplots(figsize=(8, 5))
colors = ['#2ecc71', '#e74c3c']  # Green for No, Red for Yes
bars = ax.bar(['No PCOS (0)', 'Has PCOS (1)'], class_counts.values, color=colors)

# Add count labels on bars
for bar, count, pct in zip(bars, class_counts.values, class_pct.values):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5, 
            f'{count}\n({pct}%)', ha='center', fontsize=12, fontweight='bold')

ax.set_ylabel('Number of Patients', fontsize=12)
ax.set_title('PCOS Diagnosis Distribution', fontsize=14, fontweight='bold')
ax.set_ylim(0, max(class_counts.values) * 1.15)

plt.tight_layout()
plt.show()

# Check imbalance ratio
imbalance_ratio = max(class_counts) / min(class_counts)
print(f"\nImbalance Ratio: {imbalance_ratio:.2f}:1")
if imbalance_ratio > 1.5:
    print("Dataset is imbalanced - we'll use SMOTE during training")
else:
    print("Dataset is reasonably balanced")
```

```python
# Clean Column Names (Remove Trailing/Leading Whitespace)
# Before cleaning
print("Column names BEFORE cleaning:")
print([f"'{col}'" for col in df_clean.columns[:5]])  # Show quotes to see spaces

# Clean all column names - strip whitespace
df_clean.columns = df_clean.columns.str.strip()

# After cleaning
print("\nColumn names AFTER cleaning:")
print([f"'{col}'" for col in df_clean.columns[:5]])

print("\nAll column names cleaned!")
```

```python
# Feature Categories

COLORS = {
    'pink': '#fb6f92',
    'green': '#38b000',
    'lilac': '#e0aaff',
    'amber': '#ff9e00',
    'sky': '#90e0ef',
    'peach': '#ff6d00',
}

# Target variable
TARGET = 'PCOS (Y/N)'

# Stage 1: Non-Invasive Features (symptoms, vitals, lifestyle)
STAGE1_FEATURES = [
    'Age (yrs)', 'Weight (Kg)', 'Height(Cm)', 'BMI',
    'Cycle(R/I)', 'Cycle length(days)',
    'Weight gain(Y/N)', 'hair growth(Y/N)', 'Skin darkening (Y/N)',
    'Hair loss(Y/N)', 'Pimples(Y/N)', 'Fast food (Y/N)', 'Reg.Exercise(Y/N)',
    'Marraige Status (Yrs)', 'Pregnant(Y/N)', 'No. of aborptions'
]

# Stage 2: Clinical Features (hormones, ultrasound, blood tests)
STAGE2_FEATURES = [
    'FSH(mIU/mL)', 'LH(mIU/mL)', 'FSH/LH', 'AMH(ng/mL)', 'TSH (mIU/L)',
    'Follicle No. (L)', 'Follicle No. (R)', 
    'Avg. F size (L) (mm)', 'Avg. F size (R) (mm)',
    'Endometrium (mm)', 'PRL(ng/mL)', 'Vit D3 (ng/mL)', 'PRG(ng/mL)',
    'Hb(g/dl)', 'RBS(mg/dl)'
]

# Additional features (may or may not use)
OTHER_FEATURES = [
    'Blood Group', 'Pulse rate(bpm)', 'RR (breaths/min)',
    'Hip(inch)', 'Waist(inch)', 'Waist:Hip Ratio',
    'BP _Systolic (mmHg)', 'BP _Diastolic (mmHg)'
]

print("Feature Categories Defined:\n")
print(f"Stage 1 (Non-Invasive): {len(STAGE1_FEATURES)} features")
print(f"Stage 2 (Clinical):     {len(STAGE2_FEATURES)} features")
print(f"Other Features:         {len(OTHER_FEATURES)} features")
print(f"Target:                 {TARGET}")
print(f"\nTotal:                  {len(STAGE1_FEATURES) + len(STAGE2_FEATURES) + len(OTHER_FEATURES) + 1}")
```

```python
# Correlation with PCOS Target
# Calculate correlations with target
correlations = df_clean.corr()[TARGET].drop(TARGET).sort_values(ascending=False)

# Top positive and negative correlations
print("Top 10 Features POSITIVELY Correlated with PCOS:\n")
print(correlations.head(10).to_string())

print("\nTop 10 Features NEGATIVELY Correlated with PCOS:\n")
print(correlations.tail(10).to_string())
```

```python
# Cell 16: Visualize Top Correlations with PCOS
# Get top 15 absolute correlations
top_corr = correlations.abs().sort_values(ascending=False).head(15)
top_corr_values = correlations[top_corr.index]

# Create horizontal bar chart
fig, ax = plt.subplots(figsize=(10, 8))

colors_list = [COLORS['green'] if v > 0 else COLORS['pink'] for v in top_corr_values]
bars = ax.barh(range(len(top_corr_values)), top_corr_values.values, color=colors_list)

ax.set_yticks(range(len(top_corr_values)))
ax.set_yticklabels(top_corr_values.index, fontsize=10)
ax.set_xlabel('Correlation with PCOS', fontsize=12)
ax.set_title('Top 15 Features Correlated with PCOS Diagnosis', fontsize=14, fontweight='bold')
ax.axvline(x=0, color='black', linewidth=0.5)
ax.set_facecolor('#f8f9fa')

# Add value labels
for i, (bar, val) in enumerate(zip(bars, top_corr_values.values)):
    ax.text(val + 0.02 if val > 0 else val - 0.02, i, f'{val:.3f}', 
            va='center', ha='left' if val > 0 else 'right', fontsize=9)

plt.tight_layout()
plt.gca().invert_yaxis()  # Highest correlation at top
plt.show()

print("\nInsight: Green = Positive correlation, Pink = Negative correlation")
```

```python
# Compare Feature Distributions: PCOS vs Non-PCOS
# Key features to visualize (mix of Stage 1 and Stage 2)
key_features = [
    'Follicle No. (L)', 'Follicle No. (R)', 
    'BMI', 'AMH(ng/mL)',
    'LH(mIU/mL)', 'FSH(mIU/mL)'
]

fig, axes = plt.subplots(2, 3, figsize=(14, 9))
axes = axes.flatten()

for idx, feature in enumerate(key_features):
    ax = axes[idx]
    
    # Separate data by PCOS status
    no_pcos = df_clean[df_clean[TARGET] == 0][feature]
    has_pcos = df_clean[df_clean[TARGET] == 1][feature]
    
    # Create box plots
    bp = ax.boxplot([no_pcos, has_pcos], patch_artist=True, 
                    tick_labels=['No PCOS', 'PCOS'])
    
    # Color the boxes
    bp['boxes'][0].set_facecolor(COLORS['green'])
    bp['boxes'][1].set_facecolor(COLORS['pink'])
    
    ax.set_title(feature, fontsize=11, fontweight='bold')
    ax.set_facecolor('#f8f9fa')

fig.suptitle('Feature Distributions: PCOS vs Non-PCOS', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("Insight: Higher values in pink boxes indicate features elevated in PCOS patients")
```

```python
# Binary Symptoms Comparison
binary_symptoms = [
    'Weight gain(Y/N)', 'hair growth(Y/N)', 'Skin darkening (Y/N)',
    'Hair loss(Y/N)', 'Pimples(Y/N)', 'Fast food (Y/N)', 'Reg.Exercise(Y/N)'
]

# Calculate percentage with each symptom for each group
symptom_data = []
for symptom in binary_symptoms:
    no_pcos_pct = df_clean[df_clean[TARGET] == 0][symptom].mean() * 100
    has_pcos_pct = df_clean[df_clean[TARGET] == 1][symptom].mean() * 100
    symptom_data.append({
        'Symptom': symptom.replace('(Y/N)', '').strip(),
        'No PCOS (%)': no_pcos_pct,
        'PCOS (%)': has_pcos_pct,
        'Difference': has_pcos_pct - no_pcos_pct
    })

symptom_df = pd.DataFrame(symptom_data).sort_values('Difference', ascending=False)

# Visualize
fig, ax = plt.subplots(figsize=(12, 6))

x = range(len(symptom_df))
width = 0.35

bars1 = ax.bar([i - width/2 for i in x], symptom_df['No PCOS (%)'], 
               width, label='No PCOS', color=COLORS['green'])
bars2 = ax.bar([i + width/2 for i in x], symptom_df['PCOS (%)'], 
               width, label='PCOS', color=COLORS['pink'])

ax.set_ylabel('% of Patients with Symptom', fontsize=11)
ax.set_title('Symptom Prevalence: PCOS vs Non-PCOS', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(symptom_df['Symptom'], rotation=45, ha='right')
ax.legend()
ax.set_facecolor('#f8f9fa')

plt.tight_layout()
plt.show()

print("\nSymptom Comparison Table:")
print(symptom_df.to_string(index=False))
```
